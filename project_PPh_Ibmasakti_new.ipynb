I need to clean these columns by removing the dots as thousands separators and converting them to numeric types.



```python
import pandas as pd

# ... (previous code)

# Clean numeric columns
for col in ['pendapatan', 'beban_operasional', 'penyusutan', 'nilai_perolehan', 'laba_kotor', 'laba_kena_pajak', 'pph_badan']:
    df[col] = df[col].astype(str).str.replace('.', '', regex=False)
    df[col] = pd.to_numeric(df[col], errors='coerce')

display(df.head(20))
print(df.info())
```

## Data analysis

### Subtask:
Analyze the data to identify trends, patterns, and correlations.


**Reasoning**:
Analyze trends in profit_margin and operating_expenses_ratio over time using the df_aggregated DataFrame, investigate correlations between financial variables in df, analyze distributions of key variables, and document findings.



```python
import matplotlib.pyplot as plt

# 1. Trend analysis of profit_margin and operating_expenses_ratio
plt.figure(figsize=(12, 6))
plt.plot(df_aggregated['tahun'], df_aggregated['profit_margin'], label='Profit Margin', marker='o', color='skyblue')
plt.plot(df_aggregated['tahun'], df_aggregated['operating_expenses_ratio'], label='Operating Expenses Ratio', marker='o', color='orange')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Trends in Profit Margin and Operating Expenses Ratio')
plt.legend()
plt.grid(True)
plt.show()

# Calculate year-over-year changes
df_aggregated['profit_margin_change'] = df_aggregated['profit_margin'].pct_change() * 100
df_aggregated['operating_expenses_ratio_change'] = df_aggregated['operating_expenses_ratio'].pct_change() * 100
display(df_aggregated)

# 2. Investigate correlations between financial variables
correlation_matrix = df[['pendapatan', 'laba_kotor', 'beban_operasional', 'profit_margin', 'operating_expenses_ratio']].corr()
display(correlation_matrix)
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.index)), correlation_matrix.index)
plt.title('Correlation Matrix of Financial Variables')
plt.show()

# 3. Analyze distributions of key variables
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(df['pendapatan'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Pendapatan')

plt.subplot(1, 3, 2)
plt.hist(df['laba_kotor'], bins=10, color='orange', edgecolor='black')
plt.title('Distribution of Laba Kotor')

plt.subplot(1, 3, 3)
plt.hist(df['beban_operasional'], bins=10, color='green', edgecolor='black')
plt.title('Distribution of Beban Operasional')
plt.show()

# Calculate skewness and kurtosis
skewness = df[['pendapatan', 'laba_kotor', 'beban_operasional']].skew()
kurtosis = df[['pendapatan', 'laba_kotor', 'beban_operasional']].kurt()
print("Skewness:\n", skewness)
print("\nKurtosis:\n", kurtosis)


# 4. Document findings
print("\nSummary of Analysis:")
print("Trends: Analyze the 'profit_margin_change' and 'operating_expenses_ratio_change' columns in df_aggregated to identify significant trends or fluctuations.")
print("Correlations: Examine the correlation matrix to identify strong positive or negative relationships between financial variables.")
print("Distributions: Observe the histograms and skewness/kurtosis values to understand the shape and characteristics of the key financial variable distributions. Note any significant outliers.")
```

## Data visualization

### Subtask:
Create visualizations to represent the trends, correlations, and distributions identified in the data analysis.


**Reasoning**:
Create visualizations based on the previous analysis to show trends, correlations, and distributions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Trend Visualization (using df_aggregated)
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(df_aggregated['tahun'], df_aggregated['profit_margin'], label='Profit Margin', marker='o', color='skyblue')
ax1.set_xlabel('Year')
ax1.set_ylabel('Profit Margin (%)', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.grid(True)

ax2 = ax1.twinx()
ax2.plot(df_aggregated['tahun'], df_aggregated['operating_expenses_ratio'], label='Operating Expenses Ratio', marker='x', color='orange')
ax2.set_ylabel('Operating Expenses Ratio (%)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.title('Trends in Profit Margin and Operating Expenses Ratio Over Time')
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
plt.show()

# 2. Correlation Visualization (using df)
correlation_matrix = df[['pendapatan', 'laba_kotor', 'beban_operasional', 'profit_margin', 'operating_expenses_ratio']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Financial Variables')
plt.show()

# 3. Distribution Visualization (using df)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(df['pendapatan'], bins=10, color='skyblue', edgecolor='black', kde=True, ax=axes[0])
axes[0].set_title('Distribution of Pendapatan')
axes[0].set_xlabel('Pendapatan')
axes[0].set_ylabel('Frequency')

sns.histplot(df['laba_kotor'], bins=10, color='orange', edgecolor='black', kde=True, ax=axes[1])
axes[1].set_title('Distribution of Laba Kotor')
axes[1].set_xlabel('Laba Kotor')
axes[1].set_ylabel('Frequency')


sns.histplot(df['beban_operasional'], bins=10, color='green', edgecolor='black', kde=True, ax=axes[2])
axes[2].set_title('Distribution of Beban Operasional')
axes[2].set_xlabel('Beban Operasional')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# 4. Additional Visualization (Scatter Plot Matrix using df)
selected_cols = ['pendapatan', 'laba_kotor', 'beban_operasional', 'profit_margin', 'operating_expenses_ratio', 'tahun']
sns.pairplot(df[selected_cols], hue='tahun', palette='viridis')
plt.suptitle('Scatter Plot Matrix of Financial Variables by Year', y=1.02)
plt.show()
```
