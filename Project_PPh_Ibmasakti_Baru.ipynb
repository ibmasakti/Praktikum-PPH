## Data wrangling

### Subtask:
Data Wrangling and Feature Engineering


**Reasoning**:
Convert the 'tahun' column to datetime, handle categorical variables, and engineer new features.



```python
import pandas as pd

# 1. Handle the 'tahun' column
# Since 'tahun' seems to represent the year, convert it to datetime.
# Fill any NaN values with the mode of the column.
df['tahun'] = pd.to_datetime(df['tahun'], format='%Y', errors='coerce')
df['tahun'] = df['tahun'].fillna(df['tahun'].mode()[0])

# 2. Address Categorical Variables
# 'tax_holiday_status' is a categorical variable.
# Use one-hot encoding since there's no inherent order.
df = pd.get_dummies(df, columns=['tax_holiday_status'], prefix=['tax_holiday'], drop_first=True)

# 3. Feature Engineering
# Calculate profit margin
df['profit_margin'] = (df['laba_bersih_normal'] / df['pendapatan_operasional']) * 100

# Calculate the ratio of 'ppn_keluaran' to 'ppn_masukan'
df['ppn_ratio'] = df['ppn_keluaran'] / df['ppn_masukan']

# 4. Data Type Verification
print(df.dtypes)
```

## Data analysis

### Subtask:
Perform exploratory data analysis (EDA) on the cleaned and wrangled data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Descriptive Statistics
key_numerical_vars = ['pendapatan_operasional', 'laba_sebelum_pajak', 'laba_kena_pajak', 'pph_badan_normal', 'profit_margin', 'ppn_ratio']
print("Descriptive Statistics:")
display(df[key_numerical_vars].describe())

# 2. Relationships between 'pph_badan_normal' and other variables
correlation_matrix = df[key_numerical_vars].corr()
print("\nCorrelation Matrix:")
display(correlation_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Matrix of Key Variables')
plt.show()

# Investigate relationships with scatter plots
plt.figure(figsize=(15, 10))
for i, col in enumerate(key_numerical_vars):
    if col != 'pph_badan_normal':
        plt.subplot(2, 3, i + 1)
        plt.scatter(df[col], df['pph_badan_normal'], alpha=0.7)
        plt.xlabel(col)
        plt.ylabel('pph_badan_normal')
        plt.title(f'pph_badan_normal vs. {col}')
plt.tight_layout()
plt.show()

# 3. Distribution of 'pph_badan_normal' across categories (if any)
# Assuming 'tahun' is categorical for this example
plt.figure(figsize=(10, 6))
sns.boxplot(x='tahun', y='pph_badan_normal', data=df)
plt.title('Distribution of pph_badan_normal across years')
plt.xticks(rotation=45)
plt.show()

# 4. Identify potential outliers
plt.figure(figsize=(15, 10))
for i, col in enumerate(key_numerical_vars):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```
