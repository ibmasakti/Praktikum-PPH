**Reasoning**:

```python
# Drop rows with missing values in key columns
columns_to_check = ['persediaan_awal', 'persediaan_akhir', 'hpp_rawat_inap', 'total_hpp']
df_cleaned = df.dropna(subset=columns_to_check, how='any')

# Display info of the cleaned DataFrame
display(df_cleaned.info())
display(df_cleaned.describe())
```

## Data analysis

### Subtask:
Analyze the prepared data to identify trends and patterns.


**Reasoning**:
Analyze trends in key financial metrics, explore relationships between variables, and investigate potential seasonal patterns.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Trend Analysis
plt.figure(figsize=(15, 10))

# Plotting key financial metrics
plt.subplot(3, 1, 1)
plt.plot(df['tahun'], df['pendapatan_penjualan'], label='Pendapatan Penjualan', marker='o')
plt.plot(df['tahun'], df['pendapatan_penjualan'].rolling(window=3).mean(), label='Rata-rata Bergerak 3 Tahun', linestyle='--')
plt.xlabel('Tahun')
plt.ylabel('Nilai (Juta Rupiah)')
plt.title('Tren Pendapatan Penjualan')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(df['tahun'], df['laba_bersih_normal'], label='Laba Bersih Normal', marker='o', color='green')
plt.plot(df['tahun'], df['laba_bersih_normal'].rolling(window=3).mean(), label='Rata-rata Bergerak 3 Tahun', linestyle='--', color='orange')
plt.xlabel('Tahun')
plt.ylabel('Nilai (Juta Rupiah)')
plt.title('Tren Laba Bersih Normal')
plt.legend()
plt.grid(True)


plt.subplot(3, 1, 3)
plt.plot(df['tahun'], df['profit_margin'], label='Profit Margin', marker='o', color='purple')
plt.plot(df['tahun'], df['profit_margin'].rolling(window=3).mean(), label='Rata-rata Bergerak 3 Tahun', linestyle='--', color='red')
plt.xlabel('Tahun')
plt.ylabel('Presentase')
plt.title('Tren Profit Margin')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


# 2. Correlation Analysis
correlation_matrix = df[['pendapatan_penjualan', 'laba_bersih_normal', 'profit_margin', 'operating_ratio', 'pendapatan_penjualan_growth']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasi Antar Variabel Keuangan')
plt.show()

# 3. Seasonality Analysis (Not Applicable as Data is yearly)
# Note: No time series decomposition needed due to yearly data.
print("No time series decomposition needed since the data is yearly")
```
